#Transformation

##Getting most recent login dates

```python
df['LOGIN_DATE'] = pd.to_datetime(df['LOGIN_DATE'])
```

```python
recent_dates = df.groupby('CANDIDATE_ID')['LOGIN_DATE'].max().reset_index()
df1 = df.merge(recent_dates, on=['CANDIDATE_ID', 'LOGIN_DATE'], how='inner')
df1
```

##Getting last 10 events

```python
#Sorting the values by most recent date of candidate activity
df1.sort_values(by=['CANDIDATE_ID','LOGIN_DATE'],ascending=[True,False], inplace=True)
df1
```

```python
df2 = df1[['CANDIDATE_ID','event_name']]
df2
```

```python
df2['index_within_candidate'] = df2.groupby('CANDIDATE_ID').cumcount()
df2.reset_index(drop=True, inplace=True)
```

```python
df2
```

```python
df2.sort_values(by=['CANDIDATE_ID', 'index_within_candidate'], ascending=[True, True], inplace=True)
```

```python
result = df2.groupby('CANDIDATE_ID').tail(10)
```

```python
result
```

```python
result.reset_index(drop=True, inplace=True)
result['index_within_candidate'] = result.groupby('CANDIDATE_ID').cumcount()
result
```

##Counting the number of candidates who had the same last events

```python
pivot_table = result.pivot_table(index='CANDIDATE_ID', columns='index_within_candidate', values='event_name', aggfunc='first')

pivot_table.columns = [f'event {col}' for col in pivot_table.columns]

pivot_table.reset_index(inplace=True)

pivot_table = pivot_table.fillna('')

pivot_table
```

```python
event_counts = pivot_table.groupby(['event 0', 'event 1', 'event 2', 'event 3','event 4','event 5','event 6','event 7','event 8','event 9'])['CANDIDATE_ID'].nunique().reset_index()

event_counts.columns = ['event 0', 'event 1', 'event 2', 'event 3','event 4','event 5','event 6','event 7','event 8','event 9','CANDIDATE_ID']

event_counts
```

#Visualisation

```python
import plotly.express as px
```

```python
fig = px.sunburst(event_counts, path=['event 0','event 1'], values='CANDIDATE_ID')

# Update layout for better visualization
fig.update_layout(title='Sunburst Chart of Last Events')

# Show the chart
fig.show()
```
